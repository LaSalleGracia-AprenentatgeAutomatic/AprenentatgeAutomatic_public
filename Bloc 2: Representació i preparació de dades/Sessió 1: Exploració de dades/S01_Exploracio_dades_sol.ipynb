
## ğŸ“Š EstadÃ­stica descriptiva bÃ sica
Calcula mesures centrals i de dispersiÃ³ de les variables numÃ¨riques (p. ex., `math score`, `reading score`, `writing score`).

> **Recordatori matemÃ tic (Just In Time):**
> - Mitjana: $\bar{x} = \frac{1}{n}\sum x_i$
> - DesviaciÃ³ estÃ ndard: $\sigma = \sqrt{\frac{1}{n}\sum (x_i - \bar{x})^2}$
> - Mediana: valor central de les dades ordenades.


```python

# TODO: Substitueix per les columnes numÃ¨riques del teu dataset si tenen un nom diferent
num_cols = ["math score", "reading score", "writing score"]
stats = pd.DataFrame({
    "mitjana": [data[c].mean() for c in num_cols],
    "mediana": [data[c].median() for c in num_cols],
    "desviacio_estandard": [data[c].std() for c in num_cols],
}, index=num_cols)
stats
```
